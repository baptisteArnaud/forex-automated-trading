```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/DE30_EUR.2019.M1', index_col=None, header=None)
df.columns = ['timestamp', 'Complete', 'Open', 'High', 'Low', 'Close', 'Volume']
df = df[['timestamp','Open','High','Low', 'Close', 'Volume']]
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index(['timestamp'])
df.tail()
```

```python
df_backtest = df['2019-08-06']
df_backtest
```

```python
from talib.abstract import RSI

def RSI_Yo(close):
    return RSI(close, timeperiod=14)
```

```python
from talib.abstract import BBANDS

def BB(close):
    return BBANDS(close, timeperiod=21, nbdevup=2.0, nbdevdn=2.0, matype=0)
```

```python
from backtesting import Strategy
from backtesting.lib import resample_apply


class Ichimoku(Strategy):
    
    def init(self):
        self.rsi = self.I(RSI_Yo, self.data.Close)
        self.upperband, self.middleband, self.lowerband = self.I(BB, self.data.Close)
    def next(self):
        if self.position:
            if self.position.is_long and self.data.Close >= self.middleband:
                self.position.close()
            elif self.position.is_long and self.data.Close <= self.middleband:
                self.position.close()
        else:
            if self.rsi < 30 and self.data.Close < self.lowerband:
                self.buy(sl = self.data.Close - 5)
            elif self.rsi > 70 and self.data.Close > self.upperband:
                self.sell(sl = self.data.Close + 5)
```

```python
from backtesting import Backtest

bt = Backtest(df_backtest, Ichimoku, cash=10000, commission=0)
bt.run()
```
